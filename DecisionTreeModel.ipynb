```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from PIL import Image
```

```python
train=pd.read_csv("/content/drive/Shareddrives/CMPE 255 - Team Project/train_stratified.csv",header=0)

```

```python
test=pd.read_csv("/content/drive/Shareddrives/CMPE 255 - Team Project/test_stratified.csv",header=0)
```

```python
y_train=train.iloc[:,0]
y_test=test.iloc[:,0]

x_train=train.iloc[:,1:]
x_test=test.iloc[:,1:]
```

```python
from sklearn import tree
dt_clf = tree.DecisionTreeClassifier(criterion="entropy")
dt_clf.fit(x_train, y_train)
```

```python
predictions= dt_clf.predict(x_test)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,predictions,target_names="ABCDEFGHIJKLMNOPQRSTUVWXYZ"))
```

```python
dt_clf.predict_proba(x_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test, predictions)
```

```python
dt_clf = tree.DecisionTreeClassifier(criterion="gini")
dt_clf.fit(x_train, y_train)
predictions= dt_clf.predict(x_test)
accuracy_score(y_test, predictions)
```

```python
dt_clf = tree.DecisionTreeClassifier(max_depth=4)
dt_clf.fit(x_train, y_train)
accuracy_score(y_test, predictions)
```

```python
dt_clf = tree.DecisionTreeClassifier(max_depth=3)
dt_clf.fit(x_train, y_train)
accuracy_score(y_test, predictions)
```

```python
dt_clf = tree.DecisionTreeClassifier(splitter="random")
dt_clf.fit(x_train, y_train)
accuracy_score(y_test, predictions)
```

```python
dt_clf = tree.DecisionTreeClassifier(max_depth=2)
dt_clf.fit(x_train, y_train)
accuracy_score(y_test, predictions)
```
