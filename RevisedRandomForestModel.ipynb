```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from PIL import Image
```

```python
train=pd.read_csv("/content/drive/Shareddrives/CMPE 255 - Team Project/train.csv",header=0)
test=pd.read_csv("/content/drive/Shareddrives/CMPE 255 - Team Project/test.csv",header=0)
```

```python
y_train=train.iloc[:,0]
y_test=test.iloc[:,0]

x_train=train.iloc[:,1:]
x_test=test.iloc[:,1:]
```

```python
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier()
model.fit(x_train,y_train)
```

```python
#n_estimators=100 by default
```

```python
model.score(x_test,y_test)
```

```python
model=RandomForestClassifier(n_estimators=50)
model.fit(x_train,y_train)
model.score(x_test,y_test)
```

```python
model=RandomForestClassifier(n_estimators=150)
model.fit(x_train,y_train)
model.score(x_test,y_test)
```

```python
model=RandomForestClassifier(n_estimators=120)
model_fit=model.fit(x_train,y_train)
model.score(x_test,y_test)
```

```python
model=RandomForestClassifier(n_estimators=80)
model.fit(x_train,y_train)
model.score(x_test,y_test)
```

```python
y_predicted=model.predict(x_test)
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(y_test,y_predicted))
```

```python
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(y_test,y_predicted)
np.set_printoptions(linewidth=175)
cm
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sn
plt.figure(figsize=(30,10))
sn.heatmap(cm,annot=True)
plt.xlabel('Predicted')
plt.ylabel('Truth')
```

```python
plt.figure(figsize=(30,10))
sn.heatmap(cm,annot=True,fmt='d',cmap='viridis')
plt.xlabel('Predicted')
plt.ylabel('Truth')
```
