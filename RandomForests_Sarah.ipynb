Applying Random Forests Model: (for next check-in)

```python
train=pd.read_csv("/content/drive/Shareddrives/CMPE 255 - Team Project/train.csv",header=0)
```

```python
train.head()
```

```python
target=train.iloc[:,0]
target.head()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(train.iloc[:,1:],target,test_size=0.2)
```

```python
len(x_train)
len(x_test)
len(y_train)
len(y_test)
```

```python
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier()
model.fit(x_train,y_train)
```

```python
#n_estimators=100 by default
```

```python
model.score(x_test,y_test)
```

```python
model=RandomForestClassifier(n_estimators=50)
model.fit(x_train,y_train)
model.score(x_test,y_test)
```

```python
model=RandomForestClassifier(n_estimators=150)
model.fit(x_train,y_train)
model.score(x_test,y_test)
```

```python
model=RandomForestClassifier(n_estimators=120)
model_fit=model.fit(x_train,y_train)
model.score(x_test,y_test)
```

```python
model=RandomForestClassifier(n_estimators=80)
model.fit(x_train,y_train)
model.score(x_test,y_test)
```

```python
y_predicted=model.predict(x_test)
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(y_test,y_predicted))
```

```python
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(y_test,y_predicted)
np.set_printoptions(linewidth=175)
cm
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sn
plt.figure(figsize=(30,10))
sn.heatmap(cm,annot=True)
plt.xlabel('Predicted')
plt.ylabel('Truth')
```

```python
plt.figure(figsize=(30,10))
sn.heatmap(cm,annot=True,fmt='d',cmap='viridis')
plt.xlabel('Predicted')
plt.ylabel('Truth')
```

```python
import scikitplot as skplt
```

```python
fig, ax = plt.subplots(figsize=(50, 10))
skplt.metrics.plot_confusion_matrix(y_test, y_predicted, normalize=True,ax=ax)
```
